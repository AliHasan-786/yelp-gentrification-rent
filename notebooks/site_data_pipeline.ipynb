# P1 static-site data pipeline

This notebook is the human-readable entry point for the P1 extract pipeline. The implementation lives in `scripts/build_site_data.py` so the public chart data can be regenerated non-interactively with one command.

The script validates the CSV schema, reconstructs the 30-keyword PCA score, refits the H2 metro-interaction model, validates the five metro and 2015–2022 scope, and joins the panel to the selected 2020 Census ZCTA boundaries.

```python
from pathlib import Path
import subprocess
import sys

repo_root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
subprocess.run(
    [sys.executable, repo_root / "scripts" / "build_site_data.py"],
    cwd=repo_root,
    check=True,
)
```

```python
import json

panel = json.loads((repo_root / "site" / "data" / "panel.json").read_text())
assert panel["meta"]["sourceRows"] == 1113
assert panel["meta"]["mappedRows"] == 1105
assert panel["meta"]["removedDuplicateRows"] == 8
assert panel["meta"]["zipCount"] == 238
assert panel["meta"]["years"] == list(range(2015, 2023))
panel["meta"]
```
